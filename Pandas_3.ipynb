```python
'''
Q2. Mean Income

Problem Statement:

Given a dataframe df having the income details for different individuals.

Return a Series that contains the gender-wise average income.
'''

def calc_income(df):
    '''
    INPUT: df - > dataframe

    OUTPUT: result -> dataframe
    '''
    result = df.groupby('gender', as_index=False).income.mean()
    result = result.set_index('gender')

    return result
```

```python
'''
Q3. On your marks

Problem Statement:

Given a dataframe containing student's marks for multiple subjects,

Return a dataframe which contains records of subject-wise highest marks along with their roll number.
'''

def get_marks(df):
    '''
    INPUT: df -> dataframe

    OUTPUT: result -> dataframe
    '''

    ### STEP 1: group the df using subject and get max marks

    max_marks = df.groupby('subject', as_index = False).marks.max()

    ### STEP 2: reset the index

    max_marks = max_marks.reset_index(drop=True)


    ### STEP 3: merge with original df to get roll id

    result = df.merge(max_marks).sort_values(by = ['subject'])
    result = result[['subject', 'marks', 'roll_no']]
    result = result.reset_index(drop = True)

    return result
```

```python
'''
Q6. The island of slice

Given a dataframe containing records of student and their marks, and the cutoff marks k

Return the name of students who scored marks greater than or equal to the cuttof i.e. marks >= k

'''

def filter(df, k):
    '''
    INPUT:
        df -> dataframe
        k -> cutoff (int)

    OUTPUT:
        result -> dataframe
    '''

    ### STEP 1 : Create mask

    mask = df.mask(df['marks'] < k)

    ### STEP 2: Use mask to filter rows

    result = mask.dropna()['name']

    return result
```
